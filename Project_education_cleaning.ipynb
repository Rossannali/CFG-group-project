```python
import pandas as pd 
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns # new library 
import matplotlib.pyplot as plt
import numpy as np
```

```python

data = pd.read_excel("gcse-results.xlsx" , sheet_name = '2020-21', header = [2]) 
data
```

```python
# check if we have missing values in our data

missvalues = data.shape[0] - data.dropna().shape[0]
print ("%d rows have missing values in the GCSE data" %missvalues)
```

```python
# Now let's check which columns have missing values
data.isnull().sum()
```

```python
# drop the null columns
data=data.drop(['Unnamed: 8', 'Unnamed: 11','Unnamed: 15','Unnamed: 18','Unnamed: 4'], axis=1)
data.head()
```

```python
# drop the null rows
data = data.drop(labels=[0,1,34,37,47], axis=0)
data.head()
```

```python
# rename columns to seperate the data

data.rename(columns = {'Unnamed: 2':'Number of pupils at the end of key stage 4.0',
        'Unnamed: 3':'Average Attainment 8 score per pupil.0',
        'Unnamed: 4':'Average Progress 8 score per pupil.0',
        'Percentage of pupils entered for components':'Percentage of pupils entered for components.0',
        'Percentage of pupils who achieved a strong 9-5 pass': 'Percentage of pupils who achieved a strong 9-5 pass.0',
        'Percentage of pupils who achieved a standard 9-4 pass':'Percentage of pupils who achieved a standard 9-4 pass.0', 
        'Unnamed: 9':'Number of pupils at the end of key stage 4.1',
        'Unnamed: 10':'Average Attainment 8 score per pupil.1',
        'Unnamed: 11':'Average Progress 8 score per pupil.1',
        'Unnamed: 16':'Number of pupils at the end of key stage 4.2',
        'Unnamed: 17':'Average Attainment 8 score per pupil.2',
        'Unnamed: 18':'Average Progress 8 score per pupil.2'
      }, inplace = True)
data.columns
```

#### Data in columns representing 'all' is denoted by '.0 ,  data representing 'boys' is denoted by '1 ,   data representing 'girls' is denoted by .2

```python
# final check if nulls exist
data.isnull().sum()
```

```python
## dataframe with only london boroughs
data_new = data.head(32)
data_new.head()
```

```python
          
data_new['Percentage of pupils who achieved a strong 9-5 pass.0'] = data_new['Percentage of pupils who achieved a strong 9-5 pass.0'].astype('float')    
data_new['Percentage of pupils who achieved a standard 9-4 pass.0'] = data_new['Percentage of pupils who achieved a standard 9-4 pass.0'].astype('float')

data_new['Percentage of pupils who achieved a strong 9-5 pass.1'] = data_new['Percentage of pupils who achieved a strong 9-5 pass.1'].astype('float')    
data_new['Percentage of pupils who achieved a standard 9-4 pass.1'] = data_new['Percentage of pupils who achieved a standard 9-4 pass.1'].astype('float')

data_new['Percentage of pupils who achieved a strong 9-5 pass.2'] = data_new['Percentage of pupils who achieved a strong 9-5 pass.2'].astype('float')    
data_new['Percentage of pupils who achieved a standard 9-4 pass.2'] = data_new['Percentage of pupils who achieved a standard 9-4 pass.2'].astype('float')
```

```python
data_all = data_new [['Code',   
'Area',                                                       
'Number of pupils at the end of key stage 4.0',               
'Average Attainment 8 score per pupil.0',                                           
'Percentage of pupils entered for components.0',              
'Percentage of pupils who achieved a strong 9-5 pass.0',     
'Percentage of pupils who achieved a standard 9-4 pass.0']]
data_all.head()
```

```python
data_boys = data_new [['Code',   
'Area',                                                       
'Number of pupils at the end of key stage 4.1',               
'Average Attainment 8 score per pupil.1',                                           
'Percentage of pupils entered for components.1',              
'Percentage of pupils who achieved a strong 9-5 pass.1',     
'Percentage of pupils who achieved a standard 9-4 pass.1']]
data_boys.head()
```

```python
data_boys['gender'] = 'boys'
data_boys.head()
```

```python
data_girls = data_new [['Code',   
'Area',                                                       
'Number of pupils at the end of key stage 4.2',               
'Average Attainment 8 score per pupil.2',                                           
'Percentage of pupils entered for components.2',              
'Percentage of pupils who achieved a strong 9-5 pass.2',     
'Percentage of pupils who achieved a standard 9-4 pass.2']]
data_girls['gender'] = 'girls'
data_girls.head()
```

```python
data_girls.rename(columns = {'Number of pupils at the end of key stage 4.2':'Number of pupils at the end of key stage 4.1',
        'Average Attainment 8 score per pupil.2':'Average Attainment 8 score per pupil.1',
        'Percentage of pupils entered for components.2':'Percentage of pupils entered for components.1',
        'Percentage of pupils who achieved a strong 9-5 pass.2':'Percentage of pupils who achieved a strong 9-5 pass.1',
        'Percentage of pupils who achieved a standard 9-4 pass.2':'Percentage of pupils who achieved a standard 9-4 pass.1'
      }, inplace = True)
data_girls.head()
```

```python
frames = [data_boys, data_girls]
gender_frm = pd.concat(frames,ignore_index=True)
gender_frm['Percentage of pupils who achieved a strong 9-5 pass.1']=gender_frm['Percentage of pupils who achieved a strong 9-5 pass.1'].astype('float')
gender_frm['Percentage of pupils who achieved a standard 9-4 pass.1']=gender_frm['Percentage of pupils who achieved a standard 9-4 pass.1'].astype('float')
```

```python
gender_frm
```

```python
gender_frm.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show() 
```
